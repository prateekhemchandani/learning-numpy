```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
array1=np.array([[1,2,3],[4,5,6]])
array2=np.random.rand(3,3)
array3=np.zeros((4,4))
```

```python
np.save('array1.npy', array1)
np.save('array2.npy', array2)
np.save('array3.npy', array3)
```

```python
loaded_array1 = np.load('array1.npy')
loaded_array2 = np.load('array2.npy')
loaded_array3 = np.load('array3.npy')
print("Loaded Array 1:\n", loaded_array1)
print("Loaded Array 2:\n", loaded_array2)
print("Loaded Array 3:\n", loaded_array3)
```

```python
logo=mpimg.imread('numpy_logo.png')
np.save('logo.npy', logo)
loaded_logo=np.load('logo.npy')

plt.figure(figsize=(10, 6))
plt.subplot(1,2,1)
plt.imshow(loaded_logo)
plt.title("Numpy logo")
plt.grid(False)


dark_logo = loaded_logo.copy()
dark_logo[..., :3] = 1 - loaded_logo[..., :3] 
plt.subplot(1,2,2)
plt.imshow(dark_logo)
plt.title("Dark Numpy logo")
plt.grid(False)
```
